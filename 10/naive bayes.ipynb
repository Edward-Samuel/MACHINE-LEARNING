```python
Y=data_copy[["Job Offer"]]
Y
```

```python
from sklearn.preprocessing import LabelEncoder
interactiveness_le=LabelEncoder()
X.Interactiveness=interactiveness_le.fit_transform(X["Interactiveness"])
X
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.3,random_state=0)
```

```python
def fit_gaussian_nb(X,Y):
    probab={}
    mean={}
    variance={}
    classes=np.unique(Y)
    for c in classes:
        X_c=X[Y==c]
        probab[c]=len(X_c)/len(X)
        mean[c]=np.mean(X_c,axis=0)
        variance[c]=np.var(X_c,axis=0)
    return probab,mean,variance

def predict_gaussian_nb(X, probab, mean, variance):
    predictions = []
    for x in X:
        class_probabilities = {}
        for c in probab:
            class_probabilities[c] = np.sum(
                np.log(1 / np.sqrt(2 * np.pi * variance[c])) - ((x - mean[c]) ** 2) / (2 * variance[c])) + np.log(
                probab[c])
        predictions.append(max(class_probabilities, key=class_probabilities.get))
    return predictions
```

```python
probab,mean,variance=fit_gaussian_nb(x_train,y_train.values.ravel())
y_pred=predict_gaussian_nb(x_test,probab,mean,variance)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print("CONFUSION MATRIX :\n",cm)
print("ACCURACY SCORE IS ",accuracy_score(y_test,y_pred))
```
