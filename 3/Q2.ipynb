```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
data = pd.read_csv("data.csv")
data
```

```python
X = data["Advertising_Dollars"].values.reshape(-1, 1)
Y = (data["Monthly_Sales"] > data["Monthly_Sales"].mean()).astype(int)
X,Y
```

```python
model = LogisticRegression()
model
```

```python
model.fit(X, Y) 
```

```python
x_values = np.linspace(min(data.Advertising_Dollars), max(data.Advertising_Dollars), 100).reshape(-1, 1)
y_values = model.predict_proba(x_values)[:, 1]
```

```python
plt.plot(x_values, y_values, color='green', linewidth=2)                                                                                                                                                                                                                                                                                  
plt.title("Logistic Regression ROC Curve")
plt.xlabel("Online Advertising Dollars")
plt.ylabel("Probability of Sales > Mean")
plt.show()
```
