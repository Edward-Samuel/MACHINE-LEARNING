```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
data=pd.read_csv("data.csv")
data
```

```python
plt.title("Online Advertising Dollars Vs Monthly Sales")
plt.xlabel("Online Advertising Dollars")
plt.ylabel("Monthly Sales")
plt.scatter(data.Advertising_Dollars,data.Monthly_Sales,color="blue")
```

```python
X=data.drop(["Monthly_Sales","Online Store"],axis=1)
X
```

```python
Y=data.drop(["Advertising_Dollars","Online Store"],axis=1)
Y
```

```python
model=linear_model.LinearRegression()
model
```

```python
model.fit(X,Y)
```

PREDICTION

```python
sales_predict=model.predict([[4]])
sales_predict
```

SLOPE

```python
model.coef_
```

Y-INTERCEPT

```python
model.intercept_
```

EQUATION OF LINEAR REGRESSION

```python
print("Y= MX +C")
print("Y=",model.coef_,"X +",model.intercept_)
```

```python
ans=model.coef_*4+model.intercept_
ans
```

```python
plt.scatter(data.Advertising_Dollars, data.Monthly_Sales, color="blue", marker='*')
plt.title("Online Advertising Dollars Vs Monthly Sales")
plt.xlabel("Online Advertising Dollars")
plt.ylabel("Monthly Sales")

plt.plot(data.Advertising_Dollars, model.predict(X), color='green',linewidth=2)

plt.show()
```
